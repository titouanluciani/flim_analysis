```python
## THE FLIM ANALYSIS.PY FILE
from operator import truediv
from turtle import Turtle
from typing_extensions import Self
from Reader import PTUreader
from Segmentation import Segmentation
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import deconvolve, savgol_filter, fftconvolve
from math import ceil, floor
import json
import time
from scipy.stats import sem



class FLIMdata:
    """Reading .ptu files and fitting the lifetime of a ROI."""

    def __init__(self, filename, multi_channel=False, irf_path=None, save_fit_name=None):
        self.filename = filename
        if filename[-3:] == 'ptu':
            ptu_file = PTUreader(filename, print_header_data=False)
            self.decay_data, self.intensity_image = ptu_file.get_flim_data_stack()
        # elif filename[-3:] == 'npy':
        #     self.decay_data = np.load(filename)
        else:
            print('Invalid file type.')
        # decay data axes: x, y, time

        self.tau_resolution = ptu_file.head["MeasDesc_Resolution"]*1e9

        self.irf = None
        if irf_path:
            try:
                self.irf = np.load(irf_path)
                self.bin_size = 20
                self.prep_irf()
                print('IRF loaded.')
            except FileNotFoundError:
                print('No IRF file found.')

        self.to_save = {'lifetimes': [], 'pixels_lifetimes':[],
                        'photon_counts': [],
                        'fit_range': [],
                        'exp_parameters': [],
                        # 'membrane_widths': [],
                        'internal_std': [],
                        'flim_image': [],
                       'std_error':[], 'chis':[], 'photons':[],
                       'accuracies':[], 'ph-%':[], 'chi-%':[], 'cell-num':[]}
        self.save_fit_name = save_fit_name

    def save_intensity_image(self, save_name=None):
        if not save_name:
            save_name = r"intensity_images"+self.filename[9:-3]+"png"
        plt.imsave(save_name, self.intensity_image, cmap="gray")

    def plot_whole_image(self):
        # intensity image
        fig_intensity, ax_intensity = plt.subplots()
        ax_intensity.imshow(self.intensity_image, cmap='gray')
        ax_intensity.set_title('Intensity image')

        # decay data
        tau = np.linspace(0, self.decay_data.shape[2], self.decay_data.shape[2], dtype=int)*self.tau_resolution
        y = np.sum(self.decay_data, axis=(0,1))
        photon_count = np.sum(y)

        fig_whole, ax_whole = plt.subplots(1,2)

        ax_whole[0].plot(tau, y, linewidth=1)
        try:
            ax_whole[0].plot(tau, self.irf, 'r-',linewidth=1)
        except:
            ax_whole[0].plot(tau[:len(self.irf)], self.irf, linewidth=1)
            
        ax_whole[0].set_xlabel('Time [ns]')
        ax_whole[0].set_ylabel('Intensity [counts]')
        ax_whole[0].set_title(f'TCSPC Decay (photon count = {photon_count})')

        # log plot
        ax_whole[1].plot(tau, np.log(y))
        ax_whole[1].set_xlabel('Time [ns]')
        ax_whole[1].set_ylabel('Log(Intensity) [counts]')
        ax_whole[1].set_title('TCSPC Decay on log scale')

    def make_flim_image(self, masks, membrane=True, irf=True, show_individual_plots=True, save_fit=True, plot_fits=False):
        # cut everything before peak and everything after 15 ns
        whole_hist = np.sum(self.decay_data, axis=(0,1))
        start = np.argmax(whole_hist[:int(15/self.tau_resolution)]) # don't want to choose peak at end
        end = int(15/self.tau_resolution)
        self.decay_data = self.decay_data[:,:,start:end]
        #if self.irf is not None:
            #self.tau_resolution *= self.bin_size

        # given a segmentation, fit every region and make new image

        #UNCOMMENT IF YOU WANT WHITE BACKGROUND (value = Not A Number [NAN])
        flim_image = np.ones(self.intensity_image.shape)*np.nan
        #UNCOMMENT IF YOU WANT DARK BACKGROUND (value = ZERO)
        #flim_image = np.zeros(self.intensity_image.shape)

        max_cell = int(np.amax(masks))
        #fig, ax = plt.subplots(2,2)
        
        for i in range(1, max_cell+1):#max_cell+1):
            print(f'{i}/{max_cell}')
            mask = (masks == i)

            # 1) Iterate through each pixel of mask to get the lifetime
                # loop through decay_data where mask_3d is True
            
            
                # fit lifetime on pixel
            # 2) add lifetime to flim_image or to another one to compare with summed image
            
            tau = self.fit_lifetime(mask, irf=irf, show_plot=show_individual_plots, plot_fits=plot_fits)
            print(tau)
            flim_image = np.where(mask, tau, flim_image)
            
            # Save number of photons necessary for x% accuracy with its chi-squared in order to plot at which quality of fit
            # and for which number of photons was the given accuracy obtained (save cell number too to keep track of membrane mask)
            # get index where accuracy is x%, retrieve ph_counts, chi-squared and cell number
            """
            index = self.to_save['accuracies'].index(0.05)
            self.to_save['ph-%'].append(self.to_save['photons'][index])
            self.to_save['chi-%'].append(self.to_save['chis'][index])
            self.to_save['cell-num'].append(i)"""
        '''    
        ax[0,0].set_xlabel('Lifetime (ns)')
        ax[0,0].set_ylabel('Counts')
        N = self.to_save['photons'][-1]
        ax[0,0].set_title(f'N = {N} photons')
        ax[0,0].grid(True)
        
        ax[0,1].set_xlabel("Ph counts")
        ax[0,1].set_ylabel("Lifetimes (ns)")
        ax[0,1].set_title(f'Ph counts {N} against lifetimes')
        ax[0,1].legend()
        ax[0,1].grid(True)
        
        ax[1,0].plot(self.to_save['photons'], 1 / np.sqrt(self.to_save['photons']), label='1/sqrt(N)')
        ax[1,0].set_xscale('log')
        ax[1,0].set_ylabel('Accuracies')
        ax[1,0].set_xlabel('Ph Counts')
        ax[1,0].set_title(f'N = {N} photons')
        ax[1,0].legend()
        
        ax[1,1].plot(self.to_save['photons'], 1 / np.sqrt(self.to_save['photons']),  label='1/sqrt(N)')
        ax[1,1].set_xscale('log')
        ax[1,1].set_yscale('log')
        ax[1,1].legend()    
        plt.show() '''
        sem = [np.std(self.to_save['pixels_lifetimes'][i]) / np.sqrt(len(self.to_save['pixels_lifetimes'][i])) for i in range(len(self.to_save['photon_counts']))]
        # plot lifetimes of cell membrane with errors
        plt.figure()
        plt.errorbar(self.to_save['photon_counts'], self.to_save['lifetimes'],yerr=sem,  fmt='o', capsize=5)
        plt.xlabel('Photons count')
        plt.ylabel('Lifetimes (ns)')
        plt.legend(['Mean lifetimes with sem'])
        plt.title(f'Lifetimes of cell membrane vs nb ph. mean:{np.mean(self.to_save["lifetimes"])}, std: {np.std(self.to_save["lifetimes"])}')
        plt.show()
        
        # plot flim image
        plt.figure()
        plt.imshow(flim_image)
        plt.colorbar(extend='both')
        # clim_min = floor(min(self.to_save["lifetimes"])*10)/10
        # clim_max = ceil(max(self.to_save["lifetimes"])*10)/10
        clim_min = 2.5
        clim_max = 3.2
        
        plt.clim(clim_min, clim_max)
        plt.title(f'Mean tau: {np.mean(self.to_save["lifetimes"]):.3f},'
                  f' standard deviation: {np.std(self.to_save["lifetimes"]):.3f}')

        # if save_fit:
        self.to_save['flim_image'] = flim_image.tolist()
        self.save_json()

        return flim_image

    def prep_irf(self):
        fig, ax = plt.subplots(2,1)
        x = np.linspace(0, len(self.irf), len(self.irf), dtype=int)*self.tau_resolution*self.bin_size
        ax[0].plot(x, self.irf)
        
        self.irf = self.resize_bins(y=self.irf, n=self.bin_size)

        background = self.irf[int(10/self.tau_resolution/self.bin_size):int(15/self.tau_resolution/self.bin_size)]
        bg_val = np.mean(background)
        self.irf = self.irf - bg_val

        self.irf = self.trim_data(self.irf, end_cut=23/self.bin_size)

        x = np.linspace(0, len(self.irf), len(self.irf), dtype=int)*self.tau_resolution*self.bin_size
        ax[1].plot(x, self.irf)

    def mask_data_and_sum(self, mask):
        """Set values not in mask to 0"""
        # broadcast 2d mask to 3d
        mask_3d = mask[:,:,np.newaxis]
        mask_3d = np.array(np.broadcast_to(mask_3d, np.shape(self.decay_data)), dtype=bool)

        # masked_stack = np.ma.masked_array(self.decay_data, mask=mask_3d)
        hist = np.sum(self.decay_data, axis=(0,1), where=mask_3d)
        return hist

    def mask_pixel(self, mask, fit_vars, lin_fit=False, plot_fits=True, chi_filter=True):
        """Set values not in mask to 0"""
        # convert membrane mask to boolean array
        mask = np.array(mask,dtype=bool)
        '''
        #pixels = np.where(mask_3d, self.decay_data, 0) # shape of (512, 512, 3000) for exemple
        pixels = np.ma.array(self.decay_data, mask = mask_3d)
        # # iterate through each pixel to make a fit and get lifetime, nb-ph, chi-sq
        start_cut, end_cut = self.find_range(mask, lin_fit=lin_fit)
        print('mask_pixels ; pixels ; ', pixels.shape) 
        counts = 0
        for i in range(len(pixels.compressed()) - 1): # line index, shape (512, 3000)
            #print('mask_pixels ; line and frame ; ', pixels[i,:,:])
            for j in range(len(pixels[i].compressed()) - 1): # column index shape (3000)     
                while  counts < 1:
                    #print('shape before 0-masked :',pixels[i,j,:].shape)
                    pixel = np.ma.masked_equal(pixels[i,j,:],0).compressed()
                    #print('shape after 0-masked :',pixel.shape)
                    
                    if sum(pixel) != 0: # if histogram of pixel is 0 then pass
                        print('mask_pixels ; ph counts ; ', sum(pixel))
                        y = pixel
                        print('mask_pixel histogr of pixel : ', y)
                        # find range
                        y = self.trim_data(y, start_cut=start_cut, end_cut=end_cut)
                        print('mask_pixel histogr of pixel : ', y)
                        y = pixel
                        
                        # define x (time)
                        x = np.linspace(0, len(y), len(y), dtype=int)*self.tau_resolution
                        
                        n=1
                        tau_fit, fit_vars = self.exponential_fit2(x, y, n)
                        print("tau_fit and fit_vars in mask_pixel : ", tau_fit, fit_vars)
                        chi_squared2 = self.calc_chisquared2(x, y, *fit_vars)
                        # plot
                        if show_plot:
                            self.plot_fit2(x, y, fit_vars, n, chi_squared2)
                        print(f"PIXEL chi squared 2 {n}-exponential : {chi_squared2}")
    
                        counts+=1
        '''
        shape = True
        pixels = self.decay_data[mask].shape[0]
        for k in range(pixels):
            y =  self.decay_data[mask][k,:]
            #y = np.sum(y, axis=0)
            #photon_count = np.sum(self.decay_data[mask][:k,:])
            if np.sum(self.decay_data[mask][k,:]) == 0:
                pass
            else:
                photon_count = np.sum(self.decay_data[mask][k,:])
                #n = int(len(y) / 256)
                #y = np.array([np.sum(y[i:i+n]) for i in range(0, len(y), 256)])
                x = np.linspace(0, len(y), len(y), dtype=int) * self.tau_resolution
                if shape:
                    print(y.shape)
                    shape = False
                #plt.plot(x, y, label=f'Photon counts : {np.sum(y)}')
                
                
                for n in range(1, 2):
                    tau_fit, fit_vars = self.exponential_fit2(x, y, n)
                    #print("tau_fit and fit_vars in fit_lifetime : ", tau_fit, fit_vars)
                    chi_squared2 = self.calc_chisquared2(x, y, *fit_vars)
                    # plot
                    if plot_fits:
                        self.plot_fit2(x, y, fit_vars, n, chi_squared2)
                        print(f"multi chi squared 2 {n}-exponential : {chi_squared2}")
                if chi_squared2 > 0.8 and chi_squared2 < 1.3 and chi_filter:       
                    std_error = np.std(y) / np.mean(y)
                    accuracy = std_error / np.sqrt(photon_count)

                    pixels_lifetimes.append(tau_fit)
                    chis.append(chi_squared2)
                    photons.append(photon_count)
                    std_errors.append(std_error)
                    accuracies.append(accuracy)
        
        self.to_save['pixels_lifetimes'].append(pixels_lifetimes)
        self.to_save['chis'].append(chis)
        self.to_save['photons'].append(photons)
        self.to_save['std_error'].append(std_errors)
        self.to_save['accuracies'].append(accuracies)
    
        return np.mean(pixels_lifetimes)
            
    def fit_lifetime(self, mask, lin_fit=False, membrane=True, irf=True, show_plot=True, pixels = True, plot_pixels=False, plot_fits=True):
        """"Discard the self.to_save lists when another fitting is made"""
        global pixels_lifetimes, chis, photons, std_errors, accuracies
        pixels_lifetimes, chis, photons, std_errors, accuracies = [],[],[],[],[]
        
        """Fit the lifetime for the given data"""
        # find best membrane width
        if membrane:
            # inner_w, outer_w = self.find_membrane_width(mask)
            membrane_mask = Segmentation.membrane_from_cell(mask, innerw=2, outerw=2)
            y = self.mask_data_and_sum(membrane_mask)
        else:
            y = self.mask_data_and_sum(mask)
                                      
        # decay data
        tau2 = np.linspace(0, self.decay_data.shape[2], self.decay_data.shape[2], dtype=int)*self.tau_resolution
        plt.figure()
        plt.plot(tau2, y, linewidth=1)
        
        # deconvolution with IRF
        #if irf:
        #    y = self.deconvolution(y=y)


        # find range
        start_cut, end_cut = self.find_range(membrane_mask, lin_fit=lin_fit)
        y = self.trim_data(y, start_cut=start_cut, end_cut=end_cut)

        # define x (time)
        x = np.linspace(0, len(y), len(y), dtype=int)*self.tau_resolution
        
        # decay data
        plt.plot(x, y, linewidth=1, label='processed')
        plt.legend()
        #plt.show()
        
        if lin_fit:
            tau_fit, k_fit, I0_fit = self.lin_fit(x, y)

        else:
            # exponential fit
            tau_fit1, k_fit, I0_fit = self.exponential_fit(x, y)
            if not show_plot:
                self.plot_fit(x, y, k_fit, I0_fit)
            
        #chi_squared = self.calc_chisquared(k_fit, I0_fit, x, y)
        #print(f"Chi squared : {chi_squared}")

    
        for n in range(1, 2):
            tau_fit, fit_vars = self.exponential_fit2(x, y, n)
            #print("tau_fit and fit_vars in fit_lifetime : ", tau_fit, fit_vars)
            chi_squared2 = self.calc_chisquared2(x, y, *fit_vars)
            # plot
            if show_plot:
                self.plot_fit2(x, y, fit_vars, n, chi_squared2)
            print(f"multi chi squared 2 {n}-exponential : {chi_squared2}")
        if pixels:
            tau_fit = self.mask_pixel(membrane_mask, fit_vars, plot_fits=plot_fits)
            
        if not photons: #check if list is empty (check if condition in mask pixels has never been met)
            pass
        elif plot_pixels:
            fig, ax = plt.subplots(2,2)
            
            # plot lifetimes_overall against ph_counts + accuracies against ph_counts
            ax[0,0].hist(self.to_save['pixels_lifetimes'], bins=20)
            ax[0,1].plot(self.to_save['photons'], self.to_save['pixels_lifetimes'])
            ax[1,0].plot(self.to_save['photons'], self.to_save['accuracies'], label = 'Obs')
            ax[1,1].plot(self.to_save['photons'], self.to_save['accuracies'])

            ax[0,0].set_xlabel('Lifetime (ns)')
            ax[0,0].set_ylabel('Counts')
            
            N = self.to_save['photons'][-1]
            ax[0,0].set_title(f'N = {N} photons')
            ax[0,1].set_title(f'Ph counts {N} against lifetimes')
            ax[1,0].set_title(f'N = {N} photons')
            ax[0,0].set_title(f'N = NaN photons')
            ax[0,1].set_title(f'Ph counts NaN against lifetimes')
            ax[1,0].set_title(f'N = NaN photons')
                
            ax[0,0].grid(True)
            
            ax[0,1].set_xlabel("Ph counts")
            ax[0,1].set_ylabel("Lifetimes (ns)")
            ax[0,1].legend()
            ax[0,1].grid(True)
            
            ax[1,0].plot(self.to_save['photons'], 1 / np.sqrt(self.to_save['photons']), label='1/sqrt(N)')
            ax[1,0].set_xscale('log')
            ax[1,0].set_ylabel('Accuracies')
            ax[1,0].set_xlabel('Ph Counts')
            ax[1,0].legend()
            
            ax[1,1].plot(self.to_save['photons'], 1 / np.sqrt(self.to_save['photons']),  label='1/sqrt(N)')
            ax[1,1].set_xscale('log')
            ax[1,1].set_yscale('log')
            ax[1,1].legend()    
            plt.show()
            
        photon_count = np.sum(y) # verify that same number as self.decay_data[mask]

        self.to_save['lifetimes'].append(tau_fit)
        self.to_save['photon_counts'].append(np.sum(photons)) # filtered numb of ph (self.decay_data[mask])
        self.to_save['fit_range'].append([start_cut, end_cut])
        self.to_save['exp_parameters'].append([fit_vars])
        # self.to_save['membrane_widths'].append([inner_w, outer_w])
        
        return tau_fit

    def find_membrane_width(self, mask):
        width_range = [1, 2, 3, 4]
        min_ser = 99

        for i_w in width_range:
            # print(i_w)
            membrane_mask = Segmentation.membrane_from_cell(mask, innerw=i_w, outerw=2)

            y = self.mask_data_and_sum(membrane_mask)
            x = np.linspace(0, len(y), len(y), dtype=int)*self.tau_resolution

            x, y = self.cut_optical_path(x, y, path_length=3)
            x, y = self.cut_noisy_tail(x, y, tail_length=10)

            tau_fit, k_fit, I0_fit = self.exponential_fit(x, y)

            ser = self.calc_chisquared(k_fit, I0_fit, x, y)

            if ser < min_ser:
                min_ser = ser
                best_iw = i_w
        # print('found i_w')

        min_ser = 99
        for o_w in width_range:
            membrane_mask = Segmentation.membrane_from_cell(mask, innerw=best_iw, outerw=o_w)

            y = self.mask_data_and_sum(membrane_mask)
            x = np.linspace(0, len(y), len(y), dtype=int)*self.tau_resolution

            x, y = self.cut_optical_path(x, y, path_length=3)
            x, y = self.cut_noisy_tail(x, y, tail_length=10)

            tau_fit, k_fit, I0_fit = self.exponential_fit(x, y)

            ser = self.calc_chisquared(k_fit, I0_fit, x, y)

            if ser < min_ser:
                min_ser = ser
                best_ow = o_w
        # print('found o_w')

        return best_iw, best_ow

    # def shift(self, to_shift, shift_len=2):
    #     #cut off everything beyong cut off point of 25ns and the 0th element
    #     cut_off = 25
    #     to_shift = to_shift[1:int (cut_off*self.tau_resolution)]
    #     #shift by shift_len (elements at the back are pushed to the front)
    #     shifted = np.roll(to_shift, int(shift_len/self.tau_resolution))
    #     return shifted
    
    # def prep_irf(self):
    #     shift = True
    #     if shift:
    #         self.irf = self.shift(self.irf)

    #     self.irf = self.resize_bins(y=self.irf, n=self.bin_size)

    #     background = self.irf[int(10/self.tau_resolution/self.bin_size):int(15/self.tau_resolution/self.bin_size)]
    #     bg_val = np.mean(background)
    #     self.irf = self.irf - bg_val

    #     self.irf = self.trim_data(self.irf, end_cut = 21/self.bin_size)

    #     x = np.linspace(0, len(self.irf), len(self.irf), dtype=int)*self.tau_resolution*self.bin_size
    #     fig, ax = plt.subplots()
    #     ax.plot(x, self.irf)

    def deconvolution(self,y):
        # shift = True
        # if shift:
        #     y = self.shift(y, 2)
        resized_y = self.resize_bins(y=y, n=self.bin_size)
        y_output, y_remainder = deconvolve (resized_y, self.irf)
        #y_output *= np.sum(irf)
        #plt.plot(y_output)
        return y_output

    def trim_data(self, y, start_cut=0, end_cut=0):
        start_i = int(start_cut/self.tau_resolution)
        end_i = int(end_cut/self.tau_resolution)
        y_cut = y[start_i:]
        if end_cut>0:
            y_cut = y_cut[:-end_i]
        return y_cut

    def find_range(self, membrane_mask, lin_fit=False):
        """Try different start and end cut offs.
        Choose the cut offs that give the lowest sub group standard deviation."""
        # split in 3 sub groups
        split_ys = self.split_mask(membrane_mask)
        # find start cut off
        starts = [0, 0.5, 1, 1.5, 2]
        stds = []

        for start_cut in starts:
            taus = []
            for y in split_ys:
                y_cut = self.trim_data(y, start_cut=start_cut)
                x_cut = np.linspace(0, len(y_cut), len(y_cut), dtype=int)*self.tau_resolution
                if lin_fit:
                    tau_fit, k_fit, I0_fit = self.lin_fit(x_cut, y_cut)
                else:
                    tau_fit, k_fit, I0_fit = self.exponential_fit(x_cut, y_cut)
                taus.append(tau_fit)

            stds.append(np.std(taus))
        best_start = starts[np.argmin(stds)]

        # find tail cut off
        ends = [0, 1, 2, 3, 4]
        stds = []

        for end_cut in ends:
            taus = []
            for y in split_ys:
                y_cut = self.trim_data(y, start_cut=best_start, end_cut=end_cut)
                x_cut = np.linspace(0, len(y_cut), len(y_cut), dtype=int)*self.tau_resolution
                if lin_fit:
                    tau_fit, k_fit, I0_fit = self.lin_fit(x_cut, y_cut)
                else:
                    tau_fit, k_fit, I0_fit = self.exponential_fit(x_cut, y_cut)
                taus.append(tau_fit)
            stds.append(np.std(taus))
        best_end = ends[np.argmin(stds)]
        self.to_save['internal_std'].append(min(stds))
        return best_start, best_end

    def split_mask(self, membrane_mask):
        res = []
        random_mask = np.random.randint(1, 4, membrane_mask.shape)
        split_mask = np.where(membrane_mask, random_mask, 0)

        for j in range(1, int(np.amax(split_mask))+1):
            sub_mask = (split_mask == j)
            masked_y = self.mask_data_and_sum(sub_mask)
            res.append(masked_y)
        return res

    def exponential_fit(self, x, y, print_res=False):
        k0 = 1/2.5
        I0 = np.max(y)
        fit_vars, cov = curve_fit(self.convolve_decay, x, y, p0=(k0, I0))
        k_fit, I0_fit = fit_vars
        tau_fit = 1/k_fit
        if print_res:
            print(f"tau: {tau_fit}, I0: {I0_fit}")
        return tau_fit, k_fit, I0_fit

    def exponential_fit2(self, x, y, n, print_res=False):
        k0 = 1/2.5
        try:
            I0 = np.max(y)
        except ValueError:
            I0 = y[0]
        try:
            fit_vars, cov = curve_fit(self.convolve_decay2, x, y, p0=(k0, I0)*n, maxfev= 20000)
        except (RuntimeError, TypeError) as e:
            print(e)
            fit_vars = (k0, I0)*n
            pass
        #print("Here are fit_vars : ",fit_vars)
        tau_fit = 0
        I0_fit = 0
        for i in range(len(fit_vars) // 2):
            tau_fit += fit_vars[2*i + 1] * 1/fit_vars[2*i]
            I0_fit += fit_vars[2*i + 1]
        tau_fit, I0_fit = tau_fit / I0_fit, I0_fit / n
        if print_res:
            print(f"tau: {tau_fit}, I0: {I0_fit}")
        return tau_fit, tuple(fit_vars)

    def convolve_decay(self, x, k, I0): # Convolve the exponential model with the IRF
        decay = self.exp_decay_func(x, k, I0)
        convolved_decay = fftconvolve(decay, self.irf, mode='same')
        return convolved_decay
        
    def convolve_decay2(self, x, *fit_vars): # Convolve the exponential model with the IRF
        #print("FIT VARS " , fit_vars, x)
        decay = self.exp_decay_func2(x, *fit_vars)
        convolved_decay = fftconvolve(decay, self.irf, mode='same')
        return convolved_decay
    
    @staticmethod
    def exp_decay_func(x, k, I0):
        return I0 * np.exp(-x*k)
        
    def exp_decay_func2(self, x, *fit_vars):
        #print("RIGHT AFTER FIT VARS : ", fit_vars, x)
        #x, fit_vars = fit_vars
        #print("RIGHT AFTER FIT VARS : ", fit_vars, x)
        exp = 0
        for i in range(len(fit_vars) // 2):
            exp += fit_vars[2*i + 1] * np.exp(-x*fit_vars[2*i])
        return exp

    def plot_fit(self, x, y, k, I0, bin_size=0):
        tau = 1/k
        photon_count = np.sum(y)

        fig_fit, ax_fit = plt.subplots(1, 2)
        ax_fit[0].plot(x, y, 'b-', label=f'data (photon count = {photon_count})')
        ax_fit[0].plot(x, self.convolve_decay(x, k, I0), 'r--', label=f'fit (tau = {tau:.3f}, I0 = {I0:.1f})')
        ax_fit[0].legend()
        title = f'Fit'
        if bin_size:
            title += f' (bin size = {bin_size:.2f})'
        ax_fit[0].set_title(title)
        ax_fit[0].set_xlabel('Time [ns]')
        ax_fit[0].set_ylabel('Intensity [counts]')

        # plot on log scale
        ax_fit[1].plot(x, np.log(y), 'b-', label='data')
        ax_fit[1].plot(x, np.log(self.convolve_decay(x, k, I0)), 'r--', label= 'fit')
        ax_fit[1].legend()
        ax_fit[1].set_title('Fit on log scale')
        ax_fit[1].set_xlabel('Time [ns]')
        ax_fit[1].set_ylabel('Log(intensity) [counts]')

    def plot_fit2(self, x, y, fit_vars, n, chi_squared, bin_size=0):
        #print("fit_vars in plot_fit2 : ", fit_vars, x)
        tau, I0 = 0,0
        
        for i in range(len(fit_vars) // 2):
            tau += fit_vars[2*i + 1] * 1/fit_vars[2*i]
            I0 += fit_vars[2*i + 1]
        tau, I0 = tau / I0, I0 / n
        photons_count = np.sum(y)
        y_fit = self.convolve_decay2(x, *fit_vars)
        
        # contributions of each time bin by ponderation of their respective normalized (weighted) number of photons, tau_resolution is for lifetime
        weighted_bins = 0.5 * x * y
        mean = np.sum(weighted_bins) / photons_count
        std = np.sqrt(np.sum((weighted_bins  - mean)**2) / (photons_count-1))

        fig_fit, ax_fit = plt.subplots(1, 2)
        ax_fit[0].plot(x, y, 'b-', label=f'data (photon count = {photons_count})')
        ax_fit[0].plot(x, y_fit, 'r--', label=f'fit (tau = {tau:.3f}, I0 = {I0:.1f})')
        ax_fit[0].legend()
        ax_fit[0].axvline(x=mean, color='r', linestyle='--', label="mean of hist")
        ax_fit[0].axvline(x=mean + std, color='g', linestyle='--', label="1*std")
        title = f'Fit {n}-exponential, chi = {chi_squared:.3f}'
        if bin_size:
            title += f' (bin size = {bin_size:.2f})'
        ax_fit[0].set_title(title)
        ax_fit[0].set_xlabel('Time [ns]')
        ax_fit[0].set_ylabel('Intensity [counts]')

        # plot on log scale
        """
        ax_fit[1].plot(x, np.log(y), 'b-', label='data')
        ax_fit[1].plot(x, np.log(self.convolve_decay2(x, *fit_vars)), 'r--', label= 'fit')
        ax_fit[1].legend()
        ax_fit[1].set_title('Fit on log scale')
        ax_fit[1].set_xlabel('Time [ns]')
        ax_fit[1].set_ylabel('Log(intensity) [counts]')
        """
        
        residuals = (y - y_fit) / np.sqrt(y)
        # plot of residuals
        ax_fit[1].plot(x, residuals, 'b-', label='residuals')
        ax_fit[1].plot(x, np.zeros(x.shape), 'r--', label= 'fit')
        ax_fit[1].legend()
        ax_fit[1].set_title('Residuals')
        ax_fit[1].set_xlabel('Time [ns]')
        ax_fit[1].set_ylabel('Residuals')
        fig_fit.suptitle(f'mean={mean}, std={std}')
        fig_fit.tight_layout()
    
    def resize_bins(self, y, n):
        if n == 1:
            return y
        resized_y = np.array([np.sum(y[i:i+n]) for i in range(0, len(y), n)])
        return resized_y

    def calc_chisquared(self, k_fit, I0_fit, x, y):
        # normalize
        # y_norm = y/np.sum(y)
        # chisq, p = chisquare(y_norm, self.exp_decay_func(x, k_fit, I0_fit))
        # subtract the fitted function from the histogram value, square the result,
        # divide it by the histogram value, do this for every bin and sum up all the results
        y_fit = self.convolve_decay(x, k_fit, I0_fit) 
        chisq = np.sum((y - y_fit)**2/y_fit)
        dof = np.size(x) - 2
        red_chisq = np.sqrt(chisq/dof)
        return red_chisq
        
    def calc_chisquared2(self, x, y, *fit_vars):
        # normalize
        # y_norm = y/np.sum(y)
        # chisq, p = chisquare(y_norm, self.exp_decay_func(x, k_fit, I0_fit))
        # subtract the fitted function from the histogram value, square the result,
        # divide it by the histogram value, do this for every bin and sum up all the results
        y_fit = self.convolve_decay2(x, *fit_vars) 
        chisq = np.sum((y - y_fit)**2/y_fit)
        dof = np.size(x) - 2
        red_chisq = np.sqrt(chisq/dof)
        return red_chisq

    def durbin_watson(self, x, y, *fit_vars):
        y_fit = self.convolve_decay2(x, *fit_vars) 
        r = (y - y_fit) / np.sqrt(y_fit)
        dw = np.sum((r[1:] - r[0:-1]) / np.sum(r**2))
        
    def lin_fit(self, x, y):
        # take the log of the data and do a least squares fit
        # take out values of y that are 0
        y_resized = y[y > 0]
        x_resized = x[y > 0]

        vars = np.polynomial.polynomial.polyfit(x_resized, np.log(y_resized), 1)
        # log(I0) + k*x to I0*exp(-k*x)
        k = -vars[1]
        tau = 1/k
        I0 = np.exp(vars[0])
        res = [tau, k, I0]
        # print(tau)
        return res

    def save_json(self, save_name=None):
        json_string = json.dumps(self.to_save)

        if save_name:
            with open(save_name, 'w') as outfile:
                outfile.write(json_string)

        path_split = self.filename.split('\\')

        if self.save_fit_name:
            save_path = self.save_fit_name
        elif len(path_split) == 2:
            save_path = r"saved_fits"+self.filename[9:-4]+" FLIM.json"
        else:
            try:
                save_path = r"saved_fits\\"+path_split[-2]+" "+path_split[-1][:-4]+" FLIM.json"
                with open(save_path, 'w') as outfile:
                    outfile.write(json_string)
            except:
                pass

    def calibration(self):
        return

    @staticmethod
    def open_json(json_path):
        with open(json_path) as json_file:
            data = json.load(json_file)
        return data

    @staticmethod
    def load_new_IRF(irf_load_path, irf_save_path):
        """Give the ptu file in irf_load_path.
        The histogram of the IRF is saved as a .npy file in irf_save_path."""
        irf_file = PTUreader(irf_load_path, print_header_data=False)
        decay_data, intensity_image = irf_file.get_flim_data_stack()
        print(decay_data.shape)
        # decay_data = np.sum(decay_data, axis=2)
        
        irf_hist = np.sum(np.sum(decay_data, axis=0), axis=0)
    
        print(irf_hist.shape)
        np.save(irf_save_path, irf_hist)

        x = np.linspace(0, len(irf_hist), len(irf_hist), dtype=int)*irf_file.head["MeasDesc_Resolution"]*1e9
        photon_count = np.sum(irf_hist)

        fig_whole, ax_whole = plt.subplots(1,2)

        ax_whole[0].plot(x, irf_hist, linewidth=1)
        ax_whole[0].set_xlabel('Time [ns]')
        ax_whole[0].set_ylabel('Intensity [counts]')
        ax_whole[0].set_title(f'TCSPC Decay (photon count = {photon_count})')

        # log plot
        #np.seterr(divide = 'ignore') 
        ax_whole[1].plot(x, np.log(irf_hist))
        ax_whole[1].set_xlabel('Time [ns]')
        ax_whole[1].set_ylabel('Log(Intensity) [counts]')
        ax_whole[1].set_title('TCSPC Decay on log scale')
        return irf_hist
```
